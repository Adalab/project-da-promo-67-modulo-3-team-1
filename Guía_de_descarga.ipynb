# Si el ordenador es Mac

1. Descomprime `dashboard.zip` (doble clic).
2. Abre la aplicación **Terminal** (Cmd + espacio → escribe "Terminal").
3. Escribe `cd ` (con un espacio) y ARRASTRA la carpeta descomprimida a la ventana
   de la Terminal → pulsa Enter. (Truco: así no hay que escribir la ruta.)
4. Comprueba que tiene Python: `python3 --version`
   (macOS lo trae; si no, se descarga de https://www.python.org/downloads/)
5. Instala las librerías (solo la primera vez):
   
   pip3 install -r requirements.txt

6. Arranca el dashboard:
   
   python3 app.py
   
7. Abre el navegador en: **http://127.0.0.1:8050**
8. Para cerrarlo: volver a la Terminal y pulsar `Ctrl + C`.

# Si el ordenador esWindows

1. Descomprime `dashboard.zip` (clic derecho → Extraer todo).
2. Instala Python si no lo tiene: https://www.python.org/downloads/
   IMPORTANTE: en la primera pantalla del instalador, marcar la casilla
   **"Add Python to PATH"**.
3. Abre la carpeta descomprimida, haz clic en la barra de direcciones del
   Explorador, escribe `cmd` y pulsa Enter (se abre una consola YA situada
   en esa carpeta — el mejor truco de Windows).
4. Instala las librerías (solo la primera vez):
   
   pip install -r requirements.txt
   
5. Arranca el dashboard:
   
   python app.py
   
6. Abre el navegador en: **http://127.0.0.1:8050**
7. Para cerrarlo: `Ctrl + C` en la consola.


# Problemas típicos y su solución en una línea

# Síntoma | Solución |
| `pip no se reconoce...` | Reinstalar Python marcando "Add Python to PATH" |
| `No module named dash` | Ejecutar `pip install -r requirements.txt` en la carpeta correcta |
| `FileNotFoundError: hr_limpio.csv` | El csv no está dentro de la carpeta: copiarlo junto a app.py |
| La página no carga | Comprobar que la consola dice "Dash is running" y usar http://127.0.0.1:8050 |
| El puerto 8050 está ocupado | En app.py, última línea: `app.run(debug=True, port=8051)` |
